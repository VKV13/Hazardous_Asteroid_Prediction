# outliers removing

```python
def remove_outliers(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df=df[(df>=lower_bound)&(df<=upper_bound)]
    return df
```

```python
sns.boxplot(data=Asteroid)
plt.title("Boxplot of Asteroid for outliers Checking")
plt.show()
```

```python
# Asteroid= remove_outliers(Asteroid)
Asteroid['epoch']=remove_outliers(Asteroid['epoch'])
Asteroid['q']=remove_outliers(Asteroid['q'])
Asteroid['tp']=remove_outliers(Asteroid['tp'])

sns.boxplot(data=Asteroid)
plt.title("Boxplot of Asteroid after removing outliers")
plt.show()
```

```python
# Capping
Asteroid['H'] = Asteroid['H'].clip(lower=Asteroid['H'].quantile(0.05), upper=Asteroid['H'].quantile(0.95))
Asteroid['e'] = Asteroid['e'].clip(lower=Asteroid['e'].quantile(0.05), upper=Asteroid['e'].quantile(0.95))
Asteroid['a'] = Asteroid['a'].clip(lower=Asteroid['a'].quantile(0.05), upper=Asteroid['a'].quantile(0.95))
Asteroid['i'] = Asteroid['i'].clip(lower=Asteroid['i'].quantile(0.05), upper=Asteroid['i'].quantile(0.95))
Asteroid['q'] = Asteroid['q'].clip(lower=Asteroid['q'].quantile(0.05), upper=Asteroid['q'].quantile(0.95))
Asteroid['tp'] = Asteroid['tp'].clip(lower=Asteroid['tp'].quantile(0.05), upper=Asteroid['tp'].quantile(0.95))
Asteroid['moid_ld'] = Asteroid['moid_ld'].clip(lower=Asteroid['moid_ld'].quantile(0.05), upper=Asteroid['moid_ld'].quantile(0.95))
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score

# Initialisation des modèles
models = {
    "Logistic Regression": logistic_regression,
    "Decision Tree": decision_tree,
    "SVM": svm_classifier,
    "Random Forest": random_forest,
    "KNN": knn
}
# Évaluer chaque modèle
metrics = {
    "Accuracy": {},
    "Precision": {},
    "Recall": {},
    "F1 Score": {}
}
for model_name, model in models.items():
    Y_pred = model.predict(x_test)
    metrics["Accuracy"][model_name] = accuracy_score(y_test, Y_pred)
    metrics["Precision"][model_name] = precision_score(y_test, Y_pred, average='macro')
    metrics["Recall"][model_name] = recall_score(y_test, Y_pred, average='macro')
    metrics["F1 Score"][model_name] = f1_score(y_test, Y_pred, average='macro')

# Préparer les données pour les graphiques
model_names = list(models.keys())
accuracy_scores = [metrics["Accuracy"][model] for model in model_names]
precision_scores = [metrics["Precision"][model] for model in model_names]
recall_scores = [metrics["Recall"][model] for model in model_names]
f1_scores = [metrics["F1 Score"][model] for model in model_names]

# Tracer les graphiques
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Graphique des accuracies
axes[0, 0].barh(model_names, accuracy_scores, color='skyblue')
axes[0, 0].set_xlabel('Accuracy')
axes[0, 0].set_title('Comparison of Model Accuracies')
axes[0, 0].set_xlim([0.99, 1.0])  # Limiter l'axe x pour mieux voir les différences entre les accuracy
axes[0, 0].grid(True)
for index, value in enumerate(accuracy_scores):
    axes[0, 0].text(value, index, f'{value:.4f}')

# Graphique des précisions
axes[0, 1].barh(model_names, precision_scores, color='lightgreen')
axes[0, 1].set_xlabel('Precision')
axes[0, 1].set_title('Comparison of Model Precisions')
axes[0, 1].set_xlim([0, 1.0])  # Limiter l'axe x pour mieux voir les différences
axes[0, 1].grid(True)
for index, value in enumerate(precision_scores):
    axes[0, 1].text(value, index, f'{value:.4f}')

# Graphique des recalls
axes[1, 0].barh(model_names, recall_scores, color='lightcoral')
axes[1, 0].set_xlabel('Recall')
axes[1, 0].set_title('Comparison of Model Recalls')
axes[1, 0].set_xlim([0, 1.0])  # Limiter l'axe x pour mieux voir les différences
axes[1, 0].grid(True)
for index, value in enumerate(recall_scores):
    axes[1, 0].text(value, index, f'{value:.4f}')

# Graphique des F1 scores
axes[1, 1].barh(model_names, f1_scores, color='mediumpurple')
axes[1, 1].set_xlabel('F1 Score')
axes[1, 1].set_title('Comparison of Model F1 Scores')
axes[1, 1].set_xlim([0, 1.0])  # Limiter l'axe x pour mieux voir les différences
axes[1, 1].grid(True)
for index, value in enumerate(f1_scores):
    axes[1, 1].text(value, index, f'{value:.4f}')

plt.tight_layout()
plt.show()    
```

Descision Tree has highest Recall value {To reduce False negative}

Hyper Parameter Tuning

```python
# Decision Tree
from inspect import Parameter
from sklearn.model_selection import GridSearchCV
model=DecisionTreeClassifier()
Parameter={
    'criterion': ['gini', 'entropy'],            # Split quality measure
    'max_depth': [None, 5, 10, 15, 20],           # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],              # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],                # Minimum samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2'],     # Number of features to consider for splitting
    'class_weight': [None, 'balanced']            # Weights for classes to handle class imbalance
}
grid_search = GridSearchCV(estimator=model,param_grid=Parameter,cv=5, n_jobs=-1,verbose=2)
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
best_model = grid_search.best_estimator_

test_score = best_model.score(x_test, y_test)
y_pred=best_model.predict(x_test)
print(classification_report(y_pred,y_test))
```
